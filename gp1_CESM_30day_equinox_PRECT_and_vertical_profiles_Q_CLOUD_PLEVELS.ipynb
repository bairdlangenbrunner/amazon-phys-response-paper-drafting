# import clm400 data

```python
cont_fileroot = '/beegfs/DATA/pritchard/blangenb/CESM_ARCHIVE/B_1850_CAM5.30days.cam0400.clm0400.ens_10members/ens_mean_cesm_files/'
cont_filename = 'B_1850_CAM5.30days.cam0400.clm0400.ens.cam.h0.0011-03-20-0000_0011-04-20-0000_ens_mean.nc'

cont_cam_ds = xarray.open_dataset(cont_fileroot+cont_filename, decode_times=False)
```

```python
lev_400 = cont_cam_ds['lev'].values
hyam_400 = cont_cam_ds['hyam'].values
hybm_400 = cont_cam_ds['hybm'].values
```

```python
PS_400 = cont_cam_ds['PS'].values
P0_400 = cont_cam_ds['P0'].values
```

```python
P_400 = numpy.zeros((PS_400.shape[0], hyam_400.size, PS_400.shape[1], PS_400.shape[2]))
for t_idx in range(P_400.shape[0]):
    P_400[t_idx,:,:,:] = hyam_400[:,numpy.newaxis,numpy.newaxis]*P0_400 + \
    (PS_400[t_idx,:,:]*hybm_400[:,numpy.newaxis,numpy.newaxis])
```

# Calculate latitude information

```python
lon
```

```python
lat_lo = -10.
lat_hi = 10.
lon_lo = 270.
lon_hi = 335.
```

```python
lat = cont_cam_ds['lat'].values
lon = cont_cam_ds['lon'].values

lat_indices = (lat>=lat_lo)&(lat<=lat_hi)
lon_indices = (lon>=lon_lo)&(lon<=lon_hi)

lat_subset = lat[lat_indices]
lon_subset = lon[lon_indices]
```

# take subsets

```python
P_400_subset = P_400[:,:,lat_indices,:][:,:,:,lon_indices]
```

```python
Q_400 = cont_cam_ds['Q'].sel(lat=slice(lat_lo,lat_hi),lon=slice(lon_lo,lon_hi)).values
T_400 = cont_cam_ds['T'].sel(lat=slice(lat_lo,lat_hi),lon=slice(lon_lo,lon_hi)).values
CLOUD_400 = cont_cam_ds['CLOUD'].sel(lat=slice(lat_lo,lat_hi),lon=slice(lon_lo,lon_hi)).values
PBLH_400 = cont_cam_ds['PBLH'].sel(lat=slice(lat_lo,lat_hi),lon=slice(lon_lo,lon_hi)).values
```

```python
U_400 = cont_cam_ds['U'].sel(lat=slice(lat_lo,lat_hi),lon=slice(lon_lo,lon_hi)).values
#V_400 = cont_cam_ds['v'].sel(lat=slice(lat_lo,lat_hi),lon=slice(lon_lo,lon_hi)).values
OMEGA_400 = cont_cam_ds['OMEGA'].sel(lat=slice(lat_lo,lat_hi),lon=slice(lon_lo,lon_hi)).values
```

# Convert time information

```python
time_400 = cont_cam_ds['time']
time_converted = netCDF4.num2date(time_400.values, time_400.units, time_400.calendar)
```

```python
time_manual = []
for t_idx in range(time_converted.size):
    year=time_converted[t_idx].year
    month=time_converted[t_idx].month
    day=time_converted[t_idx].day
    hour=time_converted[t_idx].hour
    minute=time_converted[t_idx].minute
    hour_round = round(hour+minute/60)
    time_manual.append(cftime.DatetimeNoLeap(year,month,day,hour_round))
```

```python
LANDFRAC = cont_cam_ds['LANDFRAC'].sel(lat=slice(lat_lo,lat_hi),lon=slice(lon_lo,lon_hi)).values[0,:,:]
```
